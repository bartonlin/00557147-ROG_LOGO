```python
!ls
path = '/kaggle/train/'

cat_filename = []
dog_filename = []

for filename in os.listdir(path):
  if filename.split('.')[0] == 'cat':
    cat_filename.append(path + filename)
  elif filename.split('.')[0] == 'dog':
    dog_filename.append(path + filename)
```

```python
import random

cat_val = random.sample(cat_filename, int(len(cat_filename) * 0.2))
dog_val = random.sample(dog_filename, int(len(dog_filename) * 0.2))

cat_train = list(set(cat_filename) - set(cat_val))
dog_train = list(set(dog_filename) - set(dog_val))
print(len(cat_train), len(cat_val))
print(len(dog_train), len(dog_val))
```

```python
import shutil

def copy_file(folder_list, target):
  for filename in folder_list:
    print(filename, filename.replace('train', target))
    shutil.copy(filename, filename.replace('train', target))

if not os.path.exists("/kaggle/dogs-vs-cats/train/cat"):
    os.makedirs("/kaggle/dogs-vs-cats/train/cat")
if not os.path.exists("/kaggle/dogs-vs-cats/train/dog"):
    os.makedirs("/kaggle/dogs-vs-cats/train/dog")

if not os.path.exists("/kaggle/dogs-vs-cats/val/cat"):
    os.makedirs("/kaggle/dogs-vs-cats/val/cat")
if not os.path.exists("/kaggle/dogs-vs-cats/val/dog"):
    os.makedirs("/kaggle/dogs-vs-cats/val/dog")

copy_file(cat_train, 'dogs-vs-cats/train/cat')
copy_file(cat_val, 'dogs-vs-cats/val/cat')
copy_file(dog_train, 'dogs-vs-cats/train/dog')
copy_file(dog_val, 'dogs-vs-cats/val/dog')
```
